# 01 - Returns and Basic Statistics

## Objectives

- Compute daily and log returns
- Estimate annualized volatility
- Visualize return distribution

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

plt.style.use("seaborn-v0_8")
```

We used adjusted prices for the SPY ETF (S&P500) from Yahoo Finance.
Adjusted close accounts for dividends and splits.

```python
#Fetch stock data
data = yf.download("SPY", start="2025-01-01")
prices = data["Close"]

print(prices.head())
```

## Returns calculations

Compute simple and log returns:

- **Simple returns**: $\ (P_t / P_{t-1}) - 1$
- **Log returns**: $\ ln(P_t / P_{t-1})$, additive over time

```python
#Simple returns

simple_returns = prices.pct_change().dropna()

#Log returns

log_returns = np.log(prices / prices.shift(1)).dropna()

print(simple_returns.head())
print(log_returns.head())
```
